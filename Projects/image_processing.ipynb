```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
```

### Load all the images for analysis

```python
# get all the filenames of the images we will analyze
image_directory = "images/"
image_files = os.listdir(image_directory)
```

```python
# load the images into a list
images = [Image.open(image_directory + image_file) for image_file in image_files]
```

```python
plt.imshow(images[-1])
```

### What is an image?
Under the hood, each image is actually an array, where each entry represents the color of a pixel! Image arrays have the shape of width x height x 3, where the third dimension contains an entry for each primary color: Red, Green, and Blue (RGB). Pixel intensity values are integers that range from 0 (completely black) - 255 (fully saturated)

```python
# let's see this:
image = images[0]
image = np.array(image)
print(image.shape)
print("max color value:", np.max(image))
print("min color value:", np.min(image))
```
